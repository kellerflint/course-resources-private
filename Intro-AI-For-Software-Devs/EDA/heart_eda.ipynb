Dataset: https://www.kaggle.com/datasets/kamilpytlak/personal-key-indicators-of-heart-disease

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart_2022_with_nans.csv')
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.dropna()

df.isnull().sum()
```

```python
# bar plot SleepHours histogram
df['SleepHours'].hist()
plt.xlabel('SleepHours')
plt.ylabel('Count')
plt.title('SleepHours Distribution')
plt.show()
```

```python
df['BMI'].hist()
plt.xlabel('BMI')
plt.ylabel('Count')
plt.title('BMI Distribution')
plt.show()
```

```python
df['HadAsthma'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('HadAsthma Distribution')
plt.show()
```

```python
df['HadDiabetes'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('HadDiabetes Distribution')
plt.show()
```

```python
df['RaceEthnicityCategory'].value_counts().plot(kind='bar')
plt.xlabel('RaceEthnicityCategory')
plt.ylabel('Count')
plt.title('RaceEthnicityCategory Distribution')
plt.show()
```

```python
df['AgeCategory'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('AgeCategory')
plt.ylabel('Count')
plt.title('AgeCategory Distribution')
plt.show()
```
